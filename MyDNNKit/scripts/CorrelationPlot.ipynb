```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
### Correlation plot on Results
resultsPath = '../Test_MixPD_MergedggF_ggH1000/llqqDNN_128_2048_4_0.003_VarSet2/'
ix_test = pd.read_pickle(resultsPath+'ResultsTestPD.pkl')
ix_train = pd.read_pickle(resultsPath+'ResultsTrainPD.pkl')

DF_Sig_Merged_ggF    = pd.concat( [ix_test.loc[(ix_test.isSignal == 1)],ix_train.loc[(ix_train.isSignal == 1)]] )
DF_Bkg_Merged_ggF    = pd.concat( [ix_test.loc[(ix_test.isSignal == 0)],ix_train.loc[(ix_train.isSignal == 0)]] )

DF_Sig_Merged_ggF  = DF_Sig_Merged_ggF[['l1_e', 'l1_pt', 'l1_eta', 'l1_phi', 'l2_e','l2_pt', 'l2_eta', 'l2_phi', 'll_m', 'll_pt','fat_jet_E', 'fat_jet_pt', 'fat_jet_eta', 'fat_jet_phi','NJETS']]
DF_Bkg_Merged_ggF  = DF_Bkg_Merged_ggF[['l1_e', 'l1_pt', 'l1_eta', 'l1_phi', 'l2_e','l2_pt', 'l2_eta', 'l2_phi', 'll_m', 'll_pt','fat_jet_E', 'fat_jet_pt', 'fat_jet_eta', 'fat_jet_phi','NJETS']]
```

```python
cmap="YlGnBu"
```

```python
plt.figure(figsize=(12,8))
corr = DF_Sig_Merged_ggF.corr()
ax = sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values,annot=True,vmin=-1.1, vmax=1.1,linewidths=0.5,fmt='0.2f',cmap=cmap)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40)
ax.set_yticklabels(ax.get_yticklabels(),rotation=0)
plt.title("Signal Parameter Correlation Matrix")
plt.savefig(resultsPath+"ParameterCorrMap_Signal.png")
```

```python
plt.figure(figsize=(12,8))
corr = DF_Bkg_Merged_ggF.corr()
ax = sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values,annot=True,vmin=-1.1, vmax=1.1,linewidths=0.5,fmt='0.2f',cmap=cmap)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40)
ax.set_yticklabels(ax.get_yticklabels(),rotation=0)
plt.title("Background Parameter Correlation Matrix")
plt.savefig(resultsPath+"ParameterCorrMap_Background.png")
```
