```python
from matplotlib import pyplot as plt
```

```python
plt.clf()
from skhep.visual import MplPlotter as skh_plt
var='l2_pt'
bins=np.linspace(min(dibosonDF[var]), max(dibosonDF[var]), 20)
plt.hist([topDF[var],dibosonDF[var],zjetsDF[var]],
             histtype='stepfilled',
             normed=False,
             bins=bins,
             weights=[ topDF['weight'],dibosonDF['weight'],zjetsDF['weight'] ],
             label=[ 'Top', 'Diboson' ,'$Z$ + jets' ],
             stacked=True)


plt.hist(signalDF[var], histtype='step', normed=False, bins=bins, weights=signalDF['weight'], label=r'ggF1000', linewidth=1, color='red', linestyle='dashed')
skh_plt.hist(dataDF[var], bins=bins, errorbars=True, histtype='marker',label='Data', color='black')

plt.legend(loc='best',prop={'size': 10})
plt.yscale('log')
plt.xlabel(var,fontsize=14)
plt.show()
```
